#### function to automate reading from main.log files

```python
def read_mains(dir):
    f=open(dir+'/main.log',"r")
    lines=f.readlines()
    #iter_txt=[]
    time_txt=[]
    energy_txt=[]
    l2_txt=[]

    for x in lines:
#    iter_txt.append(x.split(' ')[0])
        time_txt.append(x.split()[1])
        energy_txt.append(x.split()[7])    
        l2_txt.append(x.split()[10])    

    time   = [float(t) for t in time_txt]
    energy = [float(ee) for ee in energy_txt]
    l2     = [float(ee) for ee in l2_txt]
        
    f.close()
    return time, energy, l2
```

#### cases read

```python
time_dG, energy_dG, l2_dG = read_mains('dG/')
time_dG_nop, energy_dG_nop, l2_dG_nop = read_mains('rev/')
time_p, energy_p, l2_p = read_mains('p_N400/')
time_p_ppe, energy_p_ppe, l2_p_ppe = read_mains('p_PPE_N400/')
time_sp, energy_sp, l2_sp = read_mains('sp_N400/')

```
